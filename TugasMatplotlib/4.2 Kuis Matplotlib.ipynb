```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

## Soal 1
Dari data x dan y buatlah sebuah figure dengan ketentuan :
1. Judul figure adalah Data Penjualan PT. Sejahtera
2. X label adalah Pembelian
3. Y label adalah penjualan

```python
plt.plot(x, y , 'b')
plt.xlabel('Pembelian')
plt.ylabel('penjualan')


plt.title('Data Penjualan PT.Sejahtera')
plt.show()
```

## Soal 2

Buatlah sebuah figure dengan ketentuan :

2. ax1 adalah [0,0,1,1] dan ax2 adalah [0.2,0.5,.2,.2]

```python


fig = plt.figure()
ax1 = plt.axes([0,0,1,1])
ax2 = plt.axes([0.2,0.5,.2,.2])
```

sekarang isi data dari masing - masing axes tersebut dengan x dan y

```python
fig = plt.figure()
ax1 = plt.axes([0,0,1,1])
ax2 = plt.axes([0.2,0.5,.2,.2])
ax1.plot(x,y)
ax2.plot(x,y)
```

## Soal 3

Buatlah sebuah figure dengan ketentuan :
1. Menggunakan sub plot 1 baris 2 kolom

```python
ax1= plt.subplots(1,2)
```

2. Sekarang isi data plot pertama dengan data x dan y dan buat sesuai dengan tampilan di bawah
3. Sekarang isi data plot pertama dengan data x dan z dan buat sesuai dengan tampilan di bawah

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'b*')
plt.subplot(1,2,2)
plt.plot(x,z, 'b*')
```

Buat tampilannya menjadi seperti ini

```python
fig, (ax1, ax2) = plt.subplots(1,2)
fig.suptitle('Vertically stacked subplots')
ax1.plot(x, y)
ax2.plot(x, z)
```
